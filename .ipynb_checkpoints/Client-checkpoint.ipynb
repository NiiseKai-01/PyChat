```python
import socket

client = socket.socket(socket.AF_INET,socket.SOCK_STREAM)
client.connect(("0.0.0.0",55555))
# msg = client.recv(1024)
# print(msg.decode())
#client side
while True:
        y = client.recv(1024)
        print(f"Server: {y.decode()}")
        x = input("Client: ")
        client.send(x.encode())
```
